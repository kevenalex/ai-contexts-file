# Estudos sobre os dados fornecidos pelo SEART_DATA

Este notebook tem como objetivo analisar os dados, dos repositórios github, fornecidos pela ferramenta SEART.

## Importação de bibliotecas

```python
import random
import glob
import pandas as pd
import json
```

## Análise das colunas dos dados

Escolhendo JSON para análise

```python
seart_data_paths = glob.glob("../data/seart_data/*.json")

if seart_data_paths:
    SELECTED_PATH = random.choice(seart_data_paths)
    print(f"Selected file: {SELECTED_PATH}")
else:
    print("No JSON files found in the specified directory.")
```

Criação de dataframe para análise

```python
# Carregando o JSON em memória
with open(SELECTED_PATH, 'r') as f:
    data = json.load(f)

# Lidando com estrutura do JSON
if 'items' in data:
    df = pd.DataFrame(data['items'])
else:
    df = pd.DataFrame(data)
```

Descrição de cada coluna existente:

Identificação e metadados básicos:
-  id: Identificador numérico interno único do repositório no banco de dados do SEART.
-  name: Nome completo do repositório no formato usuário/nome-do-projeto.
-  isFork: Valor booleano que indica se o repositório é um fork (cópia) de outro projeto.
-  mainLanguage: A linguagem de programação principal detectada no repositório.
-  defaultBranch: O nome da branch padrão do projeto (geralmente main ou master).
-  license: O nome da licença de código aberto aplicada ao projeto (ex: MIT, Apache 2.0).
-  homepage: URL do site oficial ou documentação do projeto, se fornecida.

   Métricas de atividade e popularidade:
-  commits: Número total de commits realizados na branch padrão.
-  branches: Quantidade total de branches existentes no repositório.
-  releases: Número de versões publicadas.
-  forks: Quantidade de vezes que o repositório foi "forkado" por outros usuários.
-  watchers: Número de usuários que estão "observando" o projeto para receber notificações.
-  stargazers: Número total de estrelas que o projeto recebeu de usuários do GitHub.

   Datas e temporalidade:
-  createdAt: Data e hora em que o repositório foi criado.
-  pushedAt: Data e hora do último "push" para qualquer branch. [REVIZAR]
-  updatedAt: Data e hora da última modificação nos metadados do repositório.
-  lastCommit: Data e hora exata do commit mais recente na branch padrão.
-  lastCommitSHA: O hash do último commit realizado.

   Gestão de problemas e colaboração:
-  totalIssues: Soma total de issues criadas.
-  openIssues: Número de issues abertas.
-  totalPullRequests: Soma total de pull requests.
-  openPullRequests: Número de pull requests abertos.
-  contributors: Quantidade de usuários únicos que contribuíram com código para o projeto.

   Métricas de código:
-  blankLines: Quantidade total de linhas em branco nos arquivos de código.
-  codeLines: Quantidade total de linhas de código.
-  commentLines: Quantidade total de linhas de comentário.
-  metrics: Objeto ou lista contendo métricas detalhadas por linguagem (complexidade, linhas, etc.). [REVIZAR]
-  size: Tamanho total do repositório.

   Status e configurações:
-  hasWiki: Booleano que indica se o projeto tem a funcionalidade de Wiki ativa.
-  isArchived: Indica se o repositório foi arquivado.
-  isDisabled: Indica se o repositório foi desativado.
-  isLocked: Indica se o repositório está bloqueado.

   Listas e tags:
-  languages: Dicionário com todas as linguagens detectadas e a quantidade de bytes de cada uma.
-  labels: Lista de etiquetas utilizadas no projeto.
-  topics: Lista de palavras-chave que categorizam o projeto.
