# Estudos sobre os dados ranqueados

Este notebook tem como objetivo analisar a estrutura dos arquivos gerados pelo ranqueamento da listagem dos repositórios listados em `/data/seart_data/*`, gerando assim arquivos do tipo `ranked_DD-MM-YYYY.csv`

## Importação de bibliotecas

```python
import random
import glob
import pandas as pd
```

## Seleção de arquivo para análise

```python
ranked_data_paths = glob.glob("../data/output_rankings/*.csv")

if ranked_data_paths:
    SELECTED_PATH = random.choice(ranked_data_paths)
    print(f"Selected file: {SELECTED_PATH}")
else:
    print("No CSV files found in the output_rankings directory.")
```

Criação do dataframe

```python
df = pd.read_csv(SELECTED_PATH)
```

```python
exluded_columns = ['id', 'name', 'isFork', 'commits', 'branches', 'releases', 'forks', 'mainLanguage', 'defaultBranch', 'license', 'homepage', 'watchers', 'stargazers', 'contributors', 'size', 'createdAt', 'pushedAt', 'updatedAt', 'totalIssues', 'openIssues', 'totalPullRequests', 'openPullRequests', 'blankLines', 'codeLines', 'commentLines', 'metrics', 'lastCommit', 'lastCommitSHA', 'hasWiki', 'isArchived', 'isDisabled', 'isLocked', 'languages', 'labels', 'topics']

colunas = [col for col in df.columns if col not in exluded_columns]

df[colunas].columns.to_list()
```

## Descrição das colunas de ranqueamento [REVIZAR]

Além das colunas originais do SEART, o processo de ranqueamento adiciona métricas calculadas e normalizadas:

### Métricas de Popularidade
- `normalized_stargazers`: Número de estrelas normalizado.
- `normalized_contributors`: Número de contribuidores normalizado.
- `normalized_watchers`: Número de watchers normalizado.
- `popularity_score`: Pontuação combinada de popularidade (média ponderada das métricas normalizadas).

### Métricas de Maturidade
- `projectAge`: Idade do projeto em dias.
- `normalized_projectAge`: Idade do projeto normalizada.
- `normalized_commits`: Número de commits normalizado.
- `normalized_codeLines`: Número de linhas de código normalizado.
- `maturity_score`: Pontuação combinada de maturidade.

### Penalidades e Ajustes
- `stars_to_commits`: Razão entre estrelas e commits.
- `stars_commits_penalty`: Penalidade aplicada se a razão estrelas/commits for muito alta (indicando possível inflação artificial de estrelas).
- `text_percentage`: Porcentagem de arquivos de texto/documentação.
- `text_penalty`: Penalidade baseada na proporção de documentação vs código.
- `forks_to_stars`: Razão entre forks e estrelas.
- `forks_stars_penalty`: Penalidade baseada na razão forks/estrelas.
- `forks_to_codeLines`: Razão entre forks e linhas de código.
- `forks_codelines_penalty`: Penalidade baseada na razão forks/linhas de código (ajuda a identificar repositórios de tutoriais).
- `popularity_score_pen`: Pontuação de popularidade ajustada após a aplicação das penalidades de estrelas e forks.
- `maturity_score_pen`: Pontuação de maturidade ajustada após a aplicação das penalidades de texto e linhas de código.

### Resultado Final
- `rank_score`: Pontuação final combinada (Popularidade + Maturidade com penalidades).
- `rank`: A posição final do repositório no ranking.
